## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

- 1.1 Create a pie chart showing the count of cars by transmission type.  Display labels showing the percentage of each transmission type.  
- 1.2 Create a bar chart of cars and HP, with different colors for each car model
- 1.3 Create a scatter plot of wt (weight) vs. mpg.
- 1.4 Create a bar chart showing the average mpg for each cyl.
- 1.5 Plot a histogram of the weight column.

```python
import matplotlib.pyplot as plt

# Calculate the count of each transmission type
transmission_counts = df['am'].value_counts()

# Map the transmission type values to labels
transmission_labels = {0: 'Automatic', 1: 'Manual'}
transmission_counts.index = transmission_counts.index.map(transmission_labels)

# Create the pie chart
plt.figure(figsize=(6, 6)) # Set figure size for better visualization
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Transmission Types")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for enhanced plotting

# Bar chart of cars and HP, with different colors for each car model
plt.figure(figsize=(15, 7)) # Adjust figure size for better readability
sns.barplot(x='carmodel', y='hp', data=df, palette='tab20', hue='carmodel', dodge=False)
plt.xlabel('Car Model')
plt.ylabel('Horsepower (hp)')
plt.title('Horsepower by Car Model (with unique colors)')
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.legend([],[], frameon=False) # Remove legend as each bar is colored by its name on the x-axis
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
#Scatter plot ðŸ”§
df.plot.scatter(x='wt', y='mpg')
```

```python
import matplotlib.pyplot as plt

# Calculate the average mpg for each cylinder type
avg_mpg_by_cyl = df.groupby('cyl')['mpg'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(avg_mpg_by_cyl['cyl'], avg_mpg_by_cyl['mpg'], color='skyblue')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.title('Average MPG by Number of Cylinders')
plt.xticks(avg_mpg_by_cyl['cyl']) # Ensure all cylinder counts are shown as ticks
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Histogram ðŸ”§
df['wt'].plot(kind='hist', bins=10, title="Weight Distribution")
```
